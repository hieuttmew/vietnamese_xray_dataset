```python
import pandas as pd
```

```python
data_train = pd.read_csv('dataset_train_resized_balanced.csv')
data_test = pd.read_csv('dataset_test_resized_balanced.csv')
data_val = pd.read_csv('dataset_val_resized_balanced.csv')
```

```python
age_counts_train = data_train['Age'].value_counts().sort_index()
for age, count in age_counts_train.items():
    print(f'Age: {age} - Count: {count}')
```

```python
age_counts_test = data_test['Age'].value_counts().sort_index()
for age, count in age_counts_test.items():
    print(f'Age: {age} - Count: {count}')
```

```python
age_counts_val = data_val['Age'].value_counts().sort_index()
for age, count in age_counts_val.items():
    print(f'Age: {age} - Count: {count}')
```

```python
age_counts_train.plot(kind='bar', figsize=(20, 6), xlabel='Age - Train Dataset')
```

```python
age_counts_test.plot(kind='bar', figsize=(20, 6), xlabel='Age - Test Dataset')
```

```python
age_counts_val.plot(kind='bar', figsize=(20, 6), xlabel='Age - Validation Dataset')
```

```python
gender_counts_val.plot(kind='bar', figsize=(5, 6), xlabel='Gender - Validation Dataset')
```
