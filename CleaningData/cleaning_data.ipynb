```python
import pandas as pd

data = pd.read_csv('raw_data.csv')
data.head()
```

```python
# Duplicate extracted_text column
data['name'] = data['extracted_text']
```

```python
# Remove "TTYT-HUYEN DUC TRONG" in name column
data['name'] = data['name'].str.replace('TTYT-HUYEN DUC TRONG', '', regex=False)
data['name'] = data['name'].str.replace('UTYTHUYEN DUC TRONG', '', regex=False)
data['name'] = data['name'].str.replace('TTYTHUYE N DUC TRONG ', '', regex=False)
data['name'] = data['name'].str.replace('ITYTHUYEN DUC', '', regex=False)
data['name'] = data['name'].str.replace('TTYTHUYENDUC TRONG', '', regex=False)
data['name'] = data['name'].str.replace('YTHUYENDUC TRONG ', '', regex=False)
data['name'] = data['name'].str.replace('TTYTHOYEN DUC TRONG', '', regex=False)
data['name'] = data['name'].str.replace('TTYTEHUYENDUC TRONG', '', regex=False)

# Save a draft of cleaned data
data.to_csv('cleaned_data_draft.csv', index=False)
```

```python
# Remove all numbers from name column
data['name'] = data['name'].str.replace(r'\d+', '', regex=True)
data.to_csv('cleaned_data_draft.csv', index=False)
```

```python
# Remove "extracted_text" column
data = data.drop(columns=['extracted_text'])
data.to_csv('cleaned_data_draft.csv', index=False)
```

```python
# Remove "Pixel for Pixel" from name column
data['name'] = data['name'].str.replace('Pixel for Pixel', '', regex=False)
data.to_csv('cleaned_data_draft.csv', index=False)
```

```python
# Remove special characters from name column
data['name'] = data['name'].str.replace(r'[^\w\s]', '', regex=True)
data.to_csv('cleaned_data_draft.csv', index=False)
```

```python
# Remove " Series   Image  " from name column
data['name'] = data['name'].str.replace(' Series   Image  ', '', regex=False)
data['name'] = data['name'].str.replace(' Series    Image  ', '', regex=False)
data['name'] = data['name'].str.replace(' Series  Image ', '', regex=False)
data['name'] = data['name'].str.replace(' Series  mage ', '', regex=False)
data['name'] = data['name'].str.replace(' Series I  Image ', '', regex=False)
data['name'] = data['name'].str.replace('Series  Image  ', '', regex=False)
data['name'] = data['name'].str.replace('Series    mage  ', '', regex=False)
data.to_csv('cleaned_data_draft.csv', index=False)
```

```python
# Remove whitespace from start and end of name column
data['name'] = data['name'].str.strip()
data.to_csv('cleaned_data_draft.csv', index=False)
```
